```python
data_conll = np.loadtxt("../../../src/notebooks/Allen NLP/allen/prediction.json", dtype=str, delimiter='\n')

df_conll = pd.DataFrame(columns=['predicted', 'labels'])
for i in data_conll:
    try:
        df_conll.loc[len(df_conll)] = [json.loads(i)['predicted'], json.loads(i)['labels']]
    except:
        print(i)
```

```python
df_conll
```

```python
p = 'predicted'

pred_conll = pd.DataFrame({
      col:np.repeat(df_conll[col].values, df_conll[p].str.len())
      for col in df_conll.columns.drop(p)}
    ).assign(**{p:np.concatenate(df_conll[p].values)})[df_conll.columns]
pred_conll=pred_conll['predicted'].replace(['O', 'B-LOC', 'B-PER', 'I-PER', 'I-LOC', 'B-MISC', 'I-MISC',
       'B-ORG', 'I-ORG'],['O', 'LOCATION', 'PERSON', 'PERSON', 'LOCATION', 'MISC', 'MISC',
       'ORGANIZATION', 'ORGANIZATION'])
```

```python
t = 'labels'

true_conll = pd.DataFrame({
      col:np.repeat(df_conll[col].values, df_conll[t].str.len())
      for col in df_conll.columns.drop(t)}
    ).assign(**{t:np.concatenate(df_conll[t].values)})[df_conll.columns]
true_conll = true_conll['labels'].replace(['O', 'B-LOC', 'B-PER', 'I-PER', 'I-LOC', 'B-MISC', 'I-MISC',
       'B-ORG', 'I-ORG'],['O', 'LOCATION', 'PERSON', 'PERSON', 'LOCATION', 'MISC', 'MISC',
       'ORGANIZATION', 'ORGANIZATION'])
```

```python
pred_conll
```

```python
true_conll
```

```python
f1_score(true_conll,pred_conll,average='micro',labels=['LOCATION'])
```

```python
f1_score(true_conll,pred_conll,average='micro',labels=['PERSON'])
```

```python
f1_score(true_conll,pred_conll,average='micro',labels=['ORGANIZATION'])
```

```python
f1_score(true_conll,pred_conll,average='micro',labels=['MISC'])
```

```python
f1_score(true_conll,pred_conll,average='micro',labels=['O'])
```

```python
from allennlp.predictors import Predictor
```

```python
al = Predictor.from_path("https://storage.googleapis.com/allennlp-public-models/ner-model-2020.02.10.tar.gz")
```

```python
conll_test = open("../../../src/notebooks/Allen NLP/allen/data/test.txt").read()
split_conll_test = conll_test.split()
```

```python
# Amend class annotations to reflect Stanford's NERTagger
for n,i in enumerate(split_conll_test):
    if i == "I-PER":
        split_conll_test[n] = "PERSON"
    if i == "I-ORG":
        split_conll_test[n] = "ORGANIZATION"
    if i == "I-LOC":
        split_conll_test[n] = "LOCATION"
    if i == "B-PER":
        split_conll_test[n] = "PERSON"
    if i == "B-ORG":
        split_conll_test[n] = "ORGANIZATION"
    if i == "B-LOC":
        split_conll_test[n] = "LOCATION"
    if i == "I-MISC":
        split_conll_test[n] = "MISC"
    if i == "B-MISC":
        split_conll_test[n] = "MISC"
```

```python
conll_text = " ".join(split_conll_test[0::4])
```

```python
results = al.predict(sentence=conll_text)
```

```python
for word, tag in zip(results["words"], results["tags"]):
    print(f"{word}\t{tag}")
```

```python
unique_tags = set(results["tags"])
print(unique_tags)
```

```python
for n,i in enumerate(results["tags"]):
    if i in ["I-PER","B-PER",'U-PER','L-PER']:
        results["tags"][n] = "PERSON"
    if i in ["I-ORG","B-ORG",'U-ORG','L-ORG']:
        results["tags"][n] = "ORGANIZATION"
    if i in ["I-LOC","B-LOC",'U-LOC','L-LOC']:
        results["tags"][n] = "LOCATION"
    if i in ["I-MISC","B-MISC",'U-MISC','L-MISC']:
        results["tags"][n] = "MISC"
```

```python
unique_tags = set(results["tags"])
print(unique_tags)
```
